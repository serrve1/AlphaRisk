```python
big_label0 = []
big_label1 = []

# 각 레이블에 대한 평균값 계산
mean_by_label = df.groupby('label').mean()

# 각 컬럼에 대해 레이블 1과 레이블 0의 평균값 비교
for column in mean_by_label.columns:
    if column != 'label':  # 레이블 컬럼은 제외
        mean_label_1 = mean_by_label.loc[1, column]
        mean_label_0 = mean_by_label.loc[0, column]

        if mean_label_1 > mean_label_0:
            # print(f"레이블 1의 {column} 평균이 레이블 0의 평균보다 높습니다.")
            big_label1.append(column)

        elif mean_label_1 < mean_label_0:
            # print(f"레이블 1의 {column} 평균이 레이블 0의 평균보다 낮습니다.")
            big_label0.append(column)
        else:
            print(f"레이블 1과 레이블 0의 {column} 평균이 같습니다.")
```

```python
def cut_top_bottom(dfs, col) :
  # 특정 열의 상위 1%와 하위 1% 값을 계산
  top_1_percent = dfs[col].quantile(0.99)
  bottom_1_percent = dfs[col].quantile(0.01)

  # 상위 1%와 하위 1%를 제외한 데이터 선택
  filtered_df = dfs[(dfs[col] > bottom_1_percent) & (df[col] < top_1_percent)]
  display(filtered_df.groupby('label').mean()[col])
```

```python
def winsorize(dfs, col):
    # 특정 열의 상위 1%와 하위 1% 값을 계산
    top_1_percent = dfs[col].quantile(0.99)
    bottom_1_percent = dfs[col].quantile(0.01)

    # 상위 1%와 하위 1%를 제외한 값으로 이상치 대체
    dfs[col] = np.where(dfs[col] > top_1_percent, top_1_percent, dfs[col])
    dfs[col] = np.where(dfs[col] < bottom_1_percent, bottom_1_percent, dfs[col])

    print(dfs.groupby('label').mean()[col])
```

```python
winsorize(df, '유형자산회전율')
```

+ 윈저라이징 진행
  + 매출액증가율 (상하위 1%로 설정할 경우 정상기업 > 부실기업)
  + 자기주식은 안됨
  + 현금비율 (상하위 1%로 설정할 경우 정상기업 > 부실기업)
  + 비유동장기적합률은 안됨

+결과적으로 자기주식과 미유동장기적합률을 제외한 값은 윈저라이징 이후에 평균값이 정상기업이 부실기업보다 높게 변함
