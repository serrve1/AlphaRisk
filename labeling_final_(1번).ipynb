+ 종목코드-거래소코드, 년도-직전년도를 기준으로 머지를 진행하여 중복되는 기업을 제거하고 label df을 형성

```python
한계기업 = 상장폐지_이보배[['년도', '회사명', '거래소코드']]
상폐기업 = 상장폐지현황[['직전년도', '회사명', '종목코드']]

한계기업['년도'] = 한계기업['년도'].dt.year
상폐기업['직전년도'] = 상폐기업['직전년도'].dt.year

상폐기업.columns = 한계기업.columns
```

```python
label_df = pd.concat([한계기업, 상폐기업], axis=0).reset_index(drop=True)
print(len(한계기업))
print(len(상폐기업))
print(len(label_df))
```

```python
a = label_df[['거래소코드']]
e1 = a[a.duplicated()]['거래소코드'].tolist()
label_df[label_df['거래소코드'].isin(e1)].sort_values(['거래소코드', '년도'])
```

```python
label_df.sort_values(['거래소코드', '년도'], inplace=True)
print(len(label_df))
# label_df[label_df['거래소코드'].duplicated()].sort_values
label_df_drop_duplicate = label_df.drop_duplicates('거래소코드',keep='first')
label_df_drop_duplicate[label_df_drop_duplicate.duplicated()]
```
